# **Creating X and Y variables(features and the target)**

```python
d.columns
```

```python
x = d[['Avg. Session Length', 'Time on App', 'Time on Website','Length of Membership']].values
y = d['Yearly Amount Spent'].values
```

```python
x
```

```python
y
```

# **Splitting the dataset using train_test_split function**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

# **Training the dataset**

```python
from sklearn.linear_model import LinearRegression
z = LinearRegression()
z.fit(x_train,y_train)
```

# **Making predictions using test dataset**

```python
y_pred = z.predict(x_test)
y_pred
```

# **Predicted values Vs. Original test dataset values**

```python
sns.regplot(y_pred, y_test)
```

# **Calculating the accuracy**

```python
y_pred[1],y_test[1]
```

```python
df = pd.DataFrame({'Original Value':y_test, 'Predicted Value':y_pred})
df.head()
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_pred,y_test)
r2
```

# **Analysing the errors**

```python
from sklearn import metrics
print('Mean Absolute Error : ',metrics.mean_absolute_error(y_test,y_pred))
print('Mean Squared Error : ',metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Squared Error : ',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
